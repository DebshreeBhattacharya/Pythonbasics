## .loc    (loc is used for accessing multiple columns.)

```python
# Index a column by name same as before
# Notice that now you have to compulsarily provide a slice for the rows
df.loc[39:50,["first_name","last_name"]].head()
#df.loc[:, "first_name"]
```

```python
# Index a column by name same as before
# Notice that now you have to compulsarily provide a slice for the rows
#df.loc[50:100,["first_name","last_name"]].head()
df.loc[:, "first_name"]
```

```python
# We can index by position with the main interface - for example : 0:5 returns 50-55
sliced[0:6]
```

```python
# This will return an empty DF, as position is only upto 50
display(sliced[55:60])
```

```python
# Sometimes the original index is useful, especially if you want to modify the slice
# and assign to the original DataFrame (it automatically matches the indexes)

# But sometimes after a slice you want to reset the index
sliced=sliced.reset_index()
sliced
```

## .iloc  (The .iloc function is integer position based but it could also be used with a boolean array.)

The main purpose of using .loc and iloc is slicing of dataframe in Pandas. Function .loc is primarily used for label indexing and .iloc function is mainly applied for integer indexing.

```python
# This is good ol' numpy style indexing 
# By position on rows
# and By position on columns
df.iloc[5:10,2:5]
```

```python
df.loc[:,["first_name"]]
```
